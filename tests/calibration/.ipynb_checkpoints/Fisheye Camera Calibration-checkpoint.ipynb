```python
import cv2
assert cv2.__version__[0] >= '3', 'The fisheye module requires opencv version >= 3.0.0'
import numpy as np
import os
import glob
# from matplotlib.pyplot import imshow
from matplotlib import pyplot as plt
%matplotlib inline
```

```python
CHECKERBOARD = (6,9)
INPUT_FOLDER = "test"
OUTPUT_FOLDER = "output"
EXTENSION = "jpg"
SCALE_PERCENT = 100 # percent of original size
```

```python
subpix_criteria = (cv2.TERM_CRITERIA_EPS+cv2.TERM_CRITERIA_MAX_ITER, 30, 0.1)

calibration_flags = cv2.fisheye.CALIB_RECOMPUTE_EXTRINSIC+cv2.fisheye.CALIB_CHECK_COND+cv2.fisheye.CALIB_FIX_SKEW

objp = np.zeros((1, CHECKERBOARD[0]*CHECKERBOARD[1], 3), np.float32)

objp[0,:,:2] = np.mgrid[0:CHECKERBOARD[0], 0:CHECKERBOARD[1]].T.reshape(-1, 2)
```

```python
# input_folder = os.path.join(os.getcwd(), INPUT_FOLDER, '*.jpg')
# print(input_folder)
# images = os.listdir(input_folder)
# print(images)
images = glob.glob(os.path.join(os.getcwd(), INPUT_FOLDER, '*.%s' % EXTENSION))
print(images)
```

```python
objpoints = [] # 3d point in real world space
imgpoints = [] # 2d points in image plane.
_img_shape = None

for fname in images:
    print(fname)
    # resize image
    orig = cv2.imread(fname)
    # img = cv2.imread('/home/img/python.png', cv2.IMREAD_UNCHANGED)
    print('Original Dimensions : ',orig.shape)
    width = int(orig.shape[1] * SCALE_PERCENT / 100)
    height = int(orig.shape[0] * SCALE_PERCENT / 100)
    dim = (width, height)
    
    img = cv2.resize(orig, dim, interpolation = cv2.INTER_AREA)
 
    print('Resized Dimensions : ',img.shape)
 
    # cv2.imshow("Resized image", img)
    gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
    rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

    plt.imshow(gray, cmap='Greys_r')
    plt.title("Resized greyscale image")
    plt.show()
    
    plt.imshow(rgb)
    plt.title("Resized colour image")
    plt.show()


    if _img_shape == None:
        _img_shape = img.shape[:2]
        print(_img_shape)
    else:
        assert _img_shape == img.shape[:2], "All images must share the same size."
    # gray = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
    # Find the chess board corners
    ret, corners = cv2.findChessboardCorners(gray, CHECKERBOARD, cv2.CALIB_CB_ADAPTIVE_THRESH+cv2.CALIB_CB_FAST_CHECK+cv2.CALIB_CB_NORMALIZE_IMAGE)
    # ret, corners = cv2.findChessboardCorners(gray, CHECKERBOARD, cv2.CALIB_CB_ADAPTIVE_THRESH+cv2.CALIB_CB_NORMALIZE_IMAGE)
    # CV_CALIB_RATIONAL_MODEL, CV_CALIB_USE_INTRINSIC_GUESS, CV_CALIB_FIX_PRINCIPAL_POINT
    # ret, corners = cv2.findChessboardCorners(gray, CHECKERBOARD, +cv2.CALIB_CB_NORMALIZE_IMAGE+cv2.CALIB_RATIONAL_MODEL+cv2.CALIB_USE_INTRINSIC_GUESS+cv2.CALIB_FIX_PRINCIPAL_POINT)
    print(ret, corners)
    # If found, add object points, image points (after refining them)
    if ret == True:
        objpoints.append(objp)
        cv2.cornerSubPix(gray,corners,(3,3),(-1,-1),subpix_criteria)
        imgpoints.append(corners)
        
print(objpoints)
print(imgpoints)
```

```python
N_OK = len(objpoints)
K = np.zeros((3, 3))
D = np.zeros((4, 1))
rvecs = [np.zeros((1, 1, 3), dtype=np.float64) for i in range(N_OK)]
tvecs = [np.zeros((1, 1, 3), dtype=np.float64) for i in range(N_OK)]
```

```python
rms, _, _, _, _ = \
    cv2.fisheye.calibrate(
        objpoints,
        imgpoints,
        gray.shape[::-1],
        K,
        D,
        rvecs,
        tvecs,
        calibration_flags,
        (cv2.TERM_CRITERIA_EPS+cv2.TERM_CRITERIA_MAX_ITER, 30, 1e-6)
    )
```

```python
print("Found " + str(N_OK) + " valid images for calibration")
print("DIM=" + str(_img_shape[::-1]))
print("K=np.array(" + str(K.tolist()) + ")")
print("D=np.array(" + str(D.tolist()) + ")")
```
